```python
import tweepy
import pandas as pd
import json
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import time
from datetime import datetime
import seaborn as sns; sns.set()
import matplotlib.cm as cm
from config import consumer_key, consumer_secret, access_token, access_token_secret
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer=SentimentIntensityAnalyzer()
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
User=[]
Tweet_Body=[]
Tweet_Date=[]
Compound_List=[]
Positive_List=[]
Neutral_List=[]
Negative_List=[]
Tweet_Count=[]
```

```python
target_users = ("@BBC", "@CBS", "@CNN", "@FoxNews", "@nytimes")
```

```python
#loop through all users in target_user list
for user in target_users:
    public_tweets=api.user_timeline(user, count = 100)
    counter = 0
    
    for tweet in public_tweets:
        print(json.dumps(tweet, sort_keys=True, indent=4))
        
        #append data from tweets to empty lists
        User.append(user)
        Tweet_Body.append(tweet['text'])
        Tweet_Date.append(tweet['created_at'])
        Tweet_Count.append(counter)
        
        #run vader analysis
        Compound = analyzer.polarity_scores(tweet['text'])['compound']
        Positive = analyzer.polarity_scores(tweet['text'])['pos']
        Neutral = analyzer.polarity_scores(tweet['text'])['neu']
        Negative = analyzer.polarity_scores(tweet['text'])['neg']
        
        #append analysis results to empty lists
        Compound_List.append(Compound)
        Positive_List.append(Positive)
        Neutral_List.append(Neutral)
        Negative_List.append(Negative)

        counter =counter+1
```

```python
len(User)
```

```python
Tweet_Data = {
    "User": User,
    "Tweet": Tweet_Body,
    "Tweet Date": Tweet_Date,
    "Compound Score": Compound_List,
    "Positive Score": Positive_List,
    "Neutral Score": Neutral_List,
    "Negative Score": Negative_List,
    "Tweet Count": Tweet_Count
}
```

```python
Tweets_df=pd.DataFrame(Tweet_Data)
Tweets_df.head()
```

```python
Tweets_df.to_csv("TweetsSentimentInfo.csv")
```

```python
User_df=Tweets_df.groupby(User)
User_df.head()
```

```python
bbc_df = Tweets_df.loc[Tweets_df['User'] =='@BBC']
cbs_df = Tweets_df.loc[Tweets_df['User'] =='@CBS']
cnn_df = Tweets_df.loc[Tweets_df['User'] =='@CNN']
fox_df = Tweets_df.loc[Tweets_df['User'] =='@FoxNews']
nyt_df = Tweets_df.loc[Tweets_df['User'] =='@nytimes']

plt.scatter(bbc_df['Tweet Count'], bbc_df['Compound Score'], label = 'BBC')
plt.scatter(cbs_df['Tweet Count'], cbs_df['Compound Score'], label = 'CBS')
plt.scatter(cnn_df['Tweet Count'], cnn_df['Compound Score'], label = 'CNN')
plt.scatter(fox_df['Tweet Count'], fox_df['Compound Score'], label = 'Fox News')
plt.scatter(nyt_df['Tweet Count'], nyt_df['Compound Score'], label = 'NY Times')

plt.xlim(101, -1)
plt.title (f"Sentiment Analysis of News Outlets " + str(pd.to_datetime('today')))
plt.xlabel('Tweets Ago')
plt.ylabel('Tweet Polarity')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(True)
plt.savefig('NewsTweetCompoundSentiment.png')
```

```python
avg_sentiments_df=Tweets_df.groupby("User").mean()
avg_sentiments_df
col_of_interest=['Compound Score']
```

```python
average_compound_df = avg_sentiments_df[col_of_interest]
average_compound_df.head()
```

```python
x_vals = ['Fox News','CBS','CNN','NYT','BBC']
y_vals = average_compound_df['Compound Score']
colors = ['PLUM','THISTLE','MEDIUMORCHID','DARKORCHID','REBECCAPURPLE']

#change the fiure size
plt.figure(figsize=(10,7))

#plot the chart
plt.bar(x_vals, y_vals, color = colors, align='center', width = 1)

#labels for the axis
plt.ylabel('Tweet Polarity')
plt.title (f"Overall Media Sentiment Based on Twitter " + str(pd.to_datetime('today')))
plt.savefig('OverallMediaSentiment.png')
```

Observations:
1. It was interesting to me that at the time of pulling this data, over the last 100 tweets, CNN was the only news outlet to have a negative compound score.
2. CBS' tweets were by far the most positive - more than doubling the positive score of the next highest-ranking positive score outlet, Fox News. This is really visible in the polarity scatter plot, CBS plots are green - and hardly are seen below the 0 line. All other news outlets (besides CBS) have many observable plots on both the positive and negative score side. This could be indicative of CBS believing their audience wants to see more positive, happy, or uplifting news.
3. Even though CNN is the only news outlet with a negative compound score over thier last 100 tweets, they are closer than the 4  positive-score news outlets to being neutral.
